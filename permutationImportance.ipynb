```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import eli5


train = pd.read_csv('/Users/walkerrowe/Documents/keras/train.csv', delimiter=',')
test = pd.read_csv('/Users/walkerrowe/Documents/keras/test.csv', delimiter=',')

trainv=train.values
testv=test.values

reg = linear_model.LinearRegression()
x = trainv[0,:].reshape(-1, 1)
y = trainv[1,:].reshape(-1, 1)

model=reg.fit(x,y)

print('Coefficients: \n', reg.coef_)

xtest = testv[0,:].reshape(-1, 1)
ytest = testv[1,:].reshape(-1, 1)

preds = reg.predict(xtest)

print("Mean squared error: %.2f" % mean_squared_error(ytest,preds))

import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(reg, random_state=1).fit(x, y)
eli5.show_weights(perm)
```
